```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.iranjib.ir/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
header_div = soup.find('div', id='column2')
news_div = header_div.find_all('h3')
for i in news_div:
    print(f'news name : {i.get_text()}\nlink : {i.find('a').get('href')}')
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://news.google.com/home?hl=en-US&gl=US&ceid=US:en'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
header_div = soup.find('div', class_='Ly25Ed')
news_div = header_div.find_all('div', class_='KDoq1')
for i in news_div:
    print(f'name : {i.find('span').get_text()}\nlink : https://news.google.com/home?hl=en-US&gl=US&ceid=US:en{i.find('a').get('href')}')
```

```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.tabnak.ir/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
special_news_div_bs4 = soup.find('div', id='specialNewsH')
h2 = special_news_div_bs4.find_all('h2')
for each in h2:
    print(f'news name : {each.find('a').get_text()}\nlink : https://www.tabnak.ir{each.find('a').get('href')}')
```
